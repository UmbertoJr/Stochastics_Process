####  Linear Filters and Moving Averages
A simple method for smoothing a time series is the linear filter :

$$I_t = \sum_{j=-m}^m a_j \cdot x_{t+j}, \quad \quad$$
with $\sum_{j=-m}^m a_j = 1$

an example of filter is the **three point moving average**:
$$I_t = \frac{1}{3} (x_{t-1} + x_t + x_{t+1})$$

or the **centered four point moving average**:
$$I_t = \frac{1}{2} (\frac{x_{t-2} +x_{t-1} + x_t + x_{t+1}}{4} + \frac{x_{t-1} + x_t + x_{t+1} + x_{t+2} }{4})$$

```r
x <- Nile
a <- filter(x, 1/3 *rep(1,3))# three point moving average
b <-filter(x, 1/4 * rep(1,4))
c <- filter(x, 1/12*rep(1,12))
matplot(cbind(x,a,b,c), type = 'o')
```

## Decomposition

### Additive Decomposition Model
$$x_t = m_t + s_t + z_t$$

> **estimation** : firstly we need to estimate the trend by an appropriate moving average, then the seasonality by $\hat s_t = x_t - \hat m_t$, and average it by the right frequence. Could be also adjusted by subtracting by the mean.


### Multiplicative Decomposition Model

$$x_t = (m_t \times s_t) + z_t   $$

or: 

$$x_t = m_t \times s_t \times z_t$$

> **estimation** : firstly we need to estimate the trend by an appropriate moving average, then the seasonality by $\hat s_t = x_t / \hat m_t$, and average it by the right frequence. Could be also adjusted by diveding by their mean.

```r
CBE <- read.table('CBE.dat', header = T)
Elect.ts <- ts(CBE[,3], start = 1958, frequency = 12)

# Additive Model ----------------------------------------------------------

Elec.decom.add <-  decompose(Elect.ts)
plot(Elec.decom.add)
ts.plot(cbind(Elec.decom.add$trend, Elec.decom.add$trend + Elec.decom.add$seasonal, Elect.ts),
        col=c('blue','red','black'))
```

```r
# Multiplicative Modelling ------------------------------------------------

ELect.decom.mult <- decompose(Elect.ts, type = 'multiplicative')
plot(ELect.decom.mult)
ts.plot(cbind(ELect.decom.mult$trend, ELect.decom.mult$trend * ELect.decom.mult$seasonal, Elect.ts),
        col=c('blue','red','black'))
```

Is interesting to note that the noise is a lot less than the additive case!

## Exponential Smoothing and the Holt-Winters Method

### why Exponential

$$a_t = \alpha \cdot x_t + (1 - \alpha)\cdot a_{t-1} = \alpha (x_t - a_{t-1}) + a_{t-1}$$

so ...

$$a_t = \alpha \{ x_t + (1 - \alpha) x_{t-1} + (1 - \alpha)^2 x_{t-2} + ... + (1 - \alpha)^{t-1} x_{1} \} + (1 - \alpha)^t a_{0}$$

.

### Estimating Level, Trend and Seasonality (additive)

Level : $a_t = \alpha (x_t - s_{t-p}) + (1 + \alpha)(a_{t-1}+b_{t-1})$

Trend : $b_t = \beta (a_t - a_{t-1}) + (1 - \beta)b_{t-1}$

Seasonality : $s_t = \gamma(x_t - a_t) + (1 - \gamma)s_{t-p}$


To Forecast : $\hat x_{n + k |n} = a_n + k \cdot b_n + s_{n +k -p}$


### Estimating Level, Trend and Seasonality (multiplicative)

Level : $a_t = \alpha (x_t/s_{t-p}) + (1 + \alpha)(a_{t-1}+b_{t-1})$

Trend : $b_t = \beta (a_t - a_{t-1}) + (1 - \beta)b_{t-1}$

Seasonality : $s_t = \gamma(x_t/a_t) + (1 - \gamma)s_{t-p}$


To Forecast : $\hat x_{n + k |n} = (a_n + k \cdot b_n)\times s_{n +k -p}$


.


.

```r
data("AirPassengers")
#### additive
AP.add <- HoltWinters(AirPassengers)
plot(AP.add)


### Multiplicative
AP.mult <- HoltWinters(AirPassengers, seasonal = 'multiplicative')
plot(AP.mult)

```

```r
# additive prediction

AP.pred.add <- predict(AP.add, n.ahead = 4 * 12, prediction.interval = TRUE)
ts.plot(cbind(AirPassengers, AP.pred.add), xlab = "Time (years)",
        ylab = "Passengers (1000's)", col = c("black", "red", "blue",
                                              "blue"))
abline(v = 1961, lty = 2)


# multiplicative pred
AP.pred.mult <- predict(AP.mult, n.ahead = 4 * 12, prediction.interval = TRUE)
ts.plot(cbind(AirPassengers, AP.pred.mult), xlab = "Time (years)",
        ylab = "Passengers (1000's)", col = c("black", "red", "blue","blue"))
abline(v = 1961, lty = 2)
```

## AR model

### AR(1)

$$X_t = \alpha_1 X_{t-1} + \epsilon_t$$

$$X_t = A \alpha_1^t + (1 - \alpha_1 B)^{-1} \epsilon_t$$

$$E(X_t) = A \alpha_1^t$$

and is **asymptotically stationary** only if $|\alpha_1|\leq 1$

variance:
$$v = Var(X_t) = \frac{\sigma^2}{1 - \alpha_1^2}$$

acf:
$$\rho_1 = \alpha_1$$
$$\rho_k = \alpha_1^k$$

partial acf:
$$p_1 = \alpha_1$$
$$p_k = 0$$

### AR(p)

$$X_t = \alpha_1 X_{t-1} + \alpha_2 X_{t-2} + ... + \alpha_p X_{t-p} + \epsilon_t$$

and is **asymptotically stationary** only if the **characteristic equation** has all unit root outside the unit circle.

$$\theta(z) = 1 - \alpha_1 z - \alpha_2 z^2- ... - \alpha_p z^p$$


.



```r
eps <- rnorm(200)
x.rw <- eps
for(t in 2:200){
  x.rw[t] <- x.rw[t-1] + eps[t]
}
plot(x.rw, type = 'l')
acf(x.rw)
pacf(x.rw)
  
x.st <- eps
for(t in 2:200){
  x.st[t] <- .7*x.st[t-1] + eps[t]
}
plot(x.st, type = 'l')
acf(x.st)
pacf(x.st)
```

```r
require(tseries, quietly = T)
# Augmented Dickey-Fuller test for H0: alpha>=1
adf.test(x.rw)
adf.test(x.st)

# Phillips-Perron test 
pp.test(x.rw)
pp.test(x.st)
```

```r
x.st <- eps
x.st2 <- eps
for(t in 2:200){
    x.st[t] <- .7*x.st[t-1] + eps[t]
    x.st2[t]<- -.7*x.st2[t-1] + eps[t]
}

spectrum(x.st, log='no', span= sqrt(2*length(x.st)))
spectrum(x.st2, log='no', span= sqrt(2*length(x.st2)))
```
